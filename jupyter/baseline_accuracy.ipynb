```python
# Read the metadata
test_data = read_metadata(TEST_METADATA_PATH)
# Get the core test set
core_test_data = get_core_test_set(test_data)
# Get the paths without ext
test_paths_no_ext = get_paths_no_ext(core_test_data)
# Get the samples
test_samples = get_samples(test_paths_no_ext)
# Randomly choose 50 samples from the test set
import random
test_samples_50 = test_samples.sample(n=50, random_state=42)
# paly them and show me the spectrogram and let me predict the classes manually
import sounddevice as sd
for i in range(50):
    print(i)
    sd.play(test_samples_50.iloc[i]["wav_array"], 22050)
    sd.wait()
    print("Predicted class:")
    predicted_class = input()
    print("Correct class:")
    correct_class = test_samples_50.iloc[i]["phonetic_transcription"]
    print("Correct?")
    print(predicted_class == correct_class)
    print()
```
