I am doing a test on our previous method KNN with the sklearn.

```python
import sys

sys.path.insert(0, '..')

from utils.util import *
# Import sklearn's KNN classifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Read metadata
train_metadata = read_metadata(TRAIN_METADATA_PATH)
test_metadata = get_core_test_set(read_metadata(TEST_METADATA_PATH))
```

```python
# Get paths without extensions
train_paths_no_ext = get_paths_no_ext(train_metadata)
test_paths_no_ext = get_paths_no_ext(test_metadata)
```

```python
# Get samples
train_samples = get_samples(train_paths_no_ext)
test_samples = get_samples(test_paths_no_ext)
```

```python
# Add mfcc_vects to the training set and test set
train_samples = add_mfcc_vects(train_samples)
test_samples = add_mfcc_vects(test_samples)
```

```python
# Get X and y
X_train, y_train = get_X_y(train_samples)
X_test, y_test = get_X_y(test_samples)
```

```python
# Normalize the X matrices with z-score normalization
X_train, X_test = normalize_X(X_train, X_test)
```

```python
# Run KNN on the data
knn = KNeighborsClassifier(n_neighbors=52)

# Fit the model to the training data.
knn.fit(X_train, y_train)

# Make predictions on the test data.
y_pred = knn.predict(X_test)

# Compute the accuracy of the model.
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy: {accuracy*100:.2f}%")
```

```python
# Let the KNN model predict the top 5 phones for each given sample, so that we can calculate the top 1 - top 5 accuracy.
knn = KNeighborsClassifier(n_neighbors=111)

# Train the model
knn.fit(X_train, y_train)

# Predict the test set
y_pred = knn.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: {:.2f}'.format(accuracy))

# Calculate the top 1 - top 5 accuracy
top5_accuracy = 0
for i in range(len(y_test)):
    top5 = y_pred[i].argsort()[-5:][::-1]
    if y_test[i] in top5:
        top5_accuracy += 1
top5_accuracy /= len(y_test)
print('Top 5 Accuracy: {:.2f}'.format(top5_accuracy))
```
